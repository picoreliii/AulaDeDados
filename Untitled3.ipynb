```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Processamento dos dados
df = pd.read_csv('jogadores.csv')

# 1. Preenchimento de valores ausentes: Substituição por mediana (numéricos) e moda (categóricos)
# para manter a consistência e evitar perda de informações.
for col in ['score', 'nivel', 'idade']:
    df[col] = df[col].fillna(df[col].median())
for col in ['categoria', 'pais']:
    df[col] = df[col].fillna(df[col].mode()[0])

# 2. Correção de inconsistências em 'pais' e 'categoria': Ajuste de valores incorretos
# para garantir maior precisão nos dados.
df['pais'] = df['pais'].replace(['C#nada', 'Portuga&', '%apan'], ['Canada', 'Portugal', 'Japan'])
df['categoria'] = df['categoria'].replace(['m@go'], ['mago'])

# 3. Identificação e ajuste de valores extremos (outliers): Aplicação do método IQR
# para evitar distorções e manter a qualidade da análise.
for column in ['score', 'nivel', 'idade']:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    if column == 'score':
        df[column] = df[column].apply(lambda x: 1000000 if x > upper_bound else x)  # Limite máximo de 1000000
        df[column] = df[column].apply(lambda x: 0 if x < lower_bound else x)  # Valores negativos ajustados para 0
    elif column == 'nivel':
        df[column] = df[column].apply(lambda x: 1000 if x > upper_bound else x)  # Limite máximo de 1000
        df[column] = df[column].apply(lambda x: 0 if x < lower_bound else x)  # Valores negativos ajustados para 0
    elif column == 'idade':
        df[column] = df[column].apply(lambda x: 100 if x > upper_bound else x)  # Limite máximo de 100
        df[column] = df[column].apply(lambda x: 0 if x < lower_bound else x)  # Valores negativos ajustados para 0

# Análise exploratória: Geração de gráficos para visualizar padrões e tendências.

# Histogramas: Exibição da distribuição dos valores em cada variável.
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.hist(df['score'], bins=20)
plt.title('Distribuição de Score')
plt.xlabel('Score')
plt.ylabel('Frequência')

plt.subplot(1, 3, 2)
plt.hist(df['nivel'], bins=20)
plt.title('Distribuição de Nível')
plt.xlabel('Nível')
plt.ylabel('Frequência')

plt.subplot(1, 3, 3)
plt.hist(df['idade'], bins=20)
plt.title('Distribuição de Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()

# Box plots: Visualização da dispersão dos dados e identificação de possíveis valores extremos.
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(x=df['score'])
plt.title('Box Plot de Score')

plt.subplot(1, 3, 2)
sns.boxplot(x=df['nivel'])
plt.title('Box Plot de Nível')

plt.subplot(1, 3, 3)
sns.boxplot(x=df['idade'])
plt.title('Box Plot de Idade')

plt.tight_layout()
plt.show()

# Matriz de dispersão: Representação gráfica das relações entre as variáveis numéricas.
pd.plotting.scatter_matrix(df[['score', 'nivel', 'idade']], figsize=(10, 10))
plt.show()
```
