```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Processamento dos dados
df = pd.read_csv('jogadores.csv')

# 1. Preenchimento de valores ausentes: Substituio por mediana (numricos) e moda (categricos)
# para manter a consistncia e evitar perda de informaes.
for col in ['score', 'nivel', 'idade']:
    df[col] = df[col].fillna(df[col].median())
for col in ['categoria', 'pais']:
    df[col] = df[col].fillna(df[col].mode()[0])

# 2. Correo de inconsistncias em 'pais' e 'categoria': Ajuste de valores incorretos
# para garantir maior preciso nos dados.
df['pais'] = df['pais'].replace(['C#nada', 'Portuga&', '%apan'], ['Canada', 'Portugal', 'Japan'])
df['categoria'] = df['categoria'].replace(['m@go'], ['mago'])

# 3. Identificao e ajuste de valores extremos (outliers): Aplicao do mtodo IQR
# para evitar distores e manter a qualidade da anlise.
for column in ['score', 'nivel', 'idade']:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    if column == 'score':
        df[column] = df[column].apply(lambda x: 1000000 if x > upper_bound else x)  # Limite mximo de 1000000
        df[column] = df[column].apply(lambda x: 0 if x < lower_bound else x)  # Valores negativos ajustados para 0
    elif column == 'nivel':
        df[column] = df[column].apply(lambda x: 1000 if x > upper_bound else x)  # Limite mximo de 1000
        df[column] = df[column].apply(lambda x: 0 if x < lower_bound else x)  # Valores negativos ajustados para 0
    elif column == 'idade':
        df[column] = df[column].apply(lambda x: 100 if x > upper_bound else x)  # Limite mximo de 100
        df[column] = df[column].apply(lambda x: 0 if x < lower_bound else x)  # Valores negativos ajustados para 0

# Anlise exploratria: Gerao de grficos para visualizar padres e tendncias.

# Histogramas: Exibio da distribuio dos valores em cada varivel.
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.hist(df['score'], bins=20)
plt.title('Distribuio de Score')
plt.xlabel('Score')
plt.ylabel('Frequncia')

plt.subplot(1, 3, 2)
plt.hist(df['nivel'], bins=20)
plt.title('Distribuio de Nvel')
plt.xlabel('Nvel')
plt.ylabel('Frequncia')

plt.subplot(1, 3, 3)
plt.hist(df['idade'], bins=20)
plt.title('Distribuio de Idade')
plt.xlabel('Idade')
plt.ylabel('Frequncia')
plt.tight_layout()
plt.show()

# Box plots: Visualizao da disperso dos dados e identificao de possveis valores extremos.
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(x=df['score'])
plt.title('Box Plot de Score')

plt.subplot(1, 3, 2)
sns.boxplot(x=df['nivel'])
plt.title('Box Plot de Nvel')

plt.subplot(1, 3, 3)
sns.boxplot(x=df['idade'])
plt.title('Box Plot de Idade')

plt.tight_layout()
plt.show()

# Matriz de disperso: Representao grfica das relaes entre as variveis numricas.
pd.plotting.scatter_matrix(df[['score', 'nivel', 'idade']], figsize=(10, 10))
plt.show()
```
